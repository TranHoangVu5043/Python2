```python
s = input()
s2 = input()

d = {}

def addToDict(s: str):
    s = s[1:-1]        

    if s == "": return

    kv = s.split(", ")

    for i in kv:
        split = i.split(": ")

        k = split[0][1:-1]
        v = int(split[1])

        d[k] = d.get(k) and d[k] + v or v
    
    return d

addToDict(s)
addToDict(s2)

print(d)
```

#W6/10

```python
def calcNext(n2):
    if n2 % 2 == 0:
        return n2 / 2
    else:
        return n2 * 3 + 1

def collatzLength(n):
    if n == 1: return 0

    step = 1
    curr = n

    while curr != 1:
        curr = calcNext(curr)
        step += 1

    return step

n = int(input())

curr = 0
largest = 0

for i in range(1, n + 1):
    l = collatzLength(i)
    
    if l > largest:
        largest = l
        curr = i
print(curr, largest)
```

```python
t = list(map(int, input().split()))

d = {}

highest = False 

for i in t:
    if d.get(i): d[i] = d[i] + 1 
    else: d[i] = 1

    if not highest: highest = i

    if d[i] > d[highest]:
        highest = i 
    
    elif d[i] == d[highest]:
        highest = i <= highest and i or highest 

print(highest)
```

```python
a, b = map(int, input().split())

def gcd(a, b):
    while b > 0:
        a, b = b, a%b
 
    return a

def findAllDiv(n):
    
    s = ""

    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            s += str(i) + " "

            d = n // i
     
            if d != i:
                s += str(d) + " "
    
    s = s[:-1]
    
    return s

print(findAllDiv(gcd(a, b)))
```

#W6/32

```python
s = input()

if len(s) <= 6:
    print("Weak password")
    exit

u = 0
n = 0
special = 0

for i in s:
    if i == " ": 
        continue

    if i.isalpha():
        u += i.isupper() and 1 or 0

    elif i.isnumeric():
        n += 1
    else:
        special += 1

if (u == 0 or n == 0 or special == 0):
    print("Weak password")
    exit

print("Strong password")
```
