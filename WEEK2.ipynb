```python
import math

a, b, c = float(input("so a: ")), float(input("so b: ")), float(input("so c: "))

def isTrig (a, b, c):
    return a + b > c and a + c > b and b + c > a 

bool = isTrig(a, b, c)

if bool:
    cir = (a + b + c) / 2
    
    area = math.sqrt(cir * (cir - a) * (cir - b) * (cir - c))
    print((area))
else:
    print("Khong phai tam giac")
```

#W2/8

```python
ten = input("Tên chủ hộ: ")
cs_truoc = int(input("Chỉ số tháng trước: "))
cs_sau = int(input("Chỉ số tháng này: "))

# Tính sản lượng điện
so_dien = cs_sau - cs_truoc

if so_dien <= 0:
    print("Dữ liệu không hợp lệ!")
else:
    # Các bậc điện (giới hạn kWh, đơn giá)
    bac_dien = [
        (50, 1984),
        (50, 2050),
        (100, 2380),
        (100, 2998),
        (100, 3350),
        (float('inf'), 3460)
    ]

    tong_tien = 0
    con_lai = so_dien

    # Tính tiền theo từng bậc
    for gioi_han, gia in bac_dien:
        if con_lai > gioi_han:
            tong_tien += gioi_han * gia
            con_lai -= gioi_han
        else:
            tong_tien += con_lai * gia
            break

    # Cộng thêm 8% VAT
    tong_tien = round(tong_tien * 1.08)

    # In kết quả
    print(f"\nChủ hộ: {ten}")
    print(f"Số điện tiêu thụ: {so_dien} kWh")
    print(f"Tiền điện phải trả (đã gồm VAT 8%): {tong_tien:,} đồng")
```
