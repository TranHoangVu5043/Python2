#W11/3

```python
def happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(c) ** 2 for c in str(n))
    return n == 1

try:
    n = int(input())
    print("YES" if happy(n) else "NO")
except ValueError:
    print("ValueError")
```
