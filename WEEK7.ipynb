```python
l = list(map(int, input().split()))
n = int(input())

def binarySearch(arr, target):
    if target > arr[-1]:
        return -1 
    
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        print(mid, arr[mid])
        if arr[mid] == target:
            return mid
        elif target < arr[mid]:
            right = mid - 1
        else:
            left = mid + 1
    return -1
     
print(binarySearch(l, n))
```

#W7/3

```python
l = list(map(int, input().split()))

def bubbleSort(arr): 

    while True:  
        swapped = False

        last = len(arr) - 1
        for i in range(0, last):
            if i == last: continue
        
            if arr[i] > arr[i+1]:
                arr[i+1], arr[i] = arr[i], arr[i + 1]

                swapped = True
 
        if not swapped: break 

    return arr 

def selectionSort(arr):
    last = len(arr) - 1

    for i in range(0, last):
        minI = i

        for i2 in range(i + 1, last):
            if arr[i2] < arr[minI]:
                minI = i2

        arr[i], arr[minI] = arr[minI], arr[i]

    return arr 
print(bubbleSort(l))
print(selectionSort(l))
```

#W7/7

```python
def lis_dp(nums):
    if not nums:
        return 0
    
    n = len(nums)
    # Every element is a subsequence of length 1
    dp = [1] * n
    
    for i in range(1, n):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
                
    return max(dp)

# Example
print(lis_dp([10, 9, 2, 5, 3, 7, 101, 18]))
```

#W7/8

```python
def solve(intervals, queries):
    results = []
    
    for q in queries:
        min_len = -1

        for l, r in intervals:
            
            if l <= q <= r:
                current_len = r - l + 1
                
                if min_len == -1 or current_len < min_len:
                    min_len = current_len
        
        results.append(min_len)
        
    return results
```

#W7/9

```python
def solve(nums):
    result = []
    n = len(nums)
    
    for i in range(n):
        count = 0
        for j in range(i + 1, n):
            if nums[j] < nums[i]:
                count += 1
        result.append(count)
        
    return result

print(solve(list(map(int, input().split()))))
```

#W7/10

```python
def solve(nums, m):
    max_mod = 0
    n = len(nums)
    
    for i in range(n):
        for j in range(i, n):
            
            current_sum = sum(nums[i:j+1])
            
            max_mod = max(max_mod, current_sum % m)
    return max_mod

print(solve(list(map(int, input().split())), int(input())))
```

```python
def solve():
    sentence = input()
    keyword = input()
    # Simple brute force split and check
    words = sentence.replace('.', ' ').replace(',', ' ').split()
    count = 0
    for word in words:
        if word.lower() == keyword.lower():
            count += 1
    print(count)

solve()
```

```python
def solve():
    
    options = [input() for _ in range(4)]


    indexed_options = []
    for i in range(4):
        indexed_options.append((options[i], chr(65 + i)))
    
    indexed_options.sort()
        
    result = "".join([item[1] for item in indexed_options])
    print(" ".join(result))

solve()
```

#W7/13

```python
def solve():
    
    nums = list(map(int, input().split()))
    max_length = 0
    
    unique_nums = set(nums)
    for x in unique_nums:
        if (x + 1) in unique_nums:
            count_x = nums.count(x)
            count_x_plus_1 = nums.count(x + 1)
            max_length = max(max_length, count_x + count_x_plus_1)
            
    print(max_length)
solve()
```
