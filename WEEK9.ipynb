```python
def merge(a, b, i=0, j=0):
    if i == len(a):
        return b[j:]
    if j == len(b):
        return a[i:]
    if a[i] <= b[j]:
        return [a[i]] + merge(a, b, i + 1, j)
    return [b[j]] + merge(a, b, i, j + 1)

a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*merge(a, b))
```

#W9/2

```python
def kth(mat, n, m, k, i=0, j=0, used=None):
    if used is None:
        used = [[False]*m for _ in range(n)]
    if k == 1:
        return mat[i][j]
    used[i][j] = True
    candidates = []
    if i + 1 < n and not used[i + 1][j]:
        candidates.append((mat[i + 1][j], i + 1, j))
    if j + 1 < m and not used[i][j + 1]:
        candidates.append((mat[i][j + 1], i, j + 1))
    val, ni, nj = min(candidates)
    return kth(mat, n, m, k - 1, ni, nj, used)

n, m = map(int, input().split())
mat = [list(map(int, input().split())) for _ in range(n)]
k = int(input())
print(kth(mat, n, m, k))
```

#W9/3

```python
def solve(s, i=0, best=None):
    if best is None:
        best = s
    if i == len(s):
        return best
    for j in range(i + 1, len(s)):
        t = list(s)
        t[i], t[j] = t[j], t[i]
        if t[0] != '0':
            cand = "".join(t)
            if cand < best:
                best = cand
    return solve(s, i + 1, best)

s = input().strip()
print(solve(s))
```

```python
def can_convert(a, b):
    i = j = 0
    while i < len(a) and j < len(b):
        if a[i].upper() == b[j]:
            j += 1
        i += 1
    return j == len(b)

q = int(input())
for _ in range(q):
    a = input().strip()
    b = input().strip()
    print("YES" if can_convert(a, b) else "NO")
```

#W9/5

```python
def is_subarray(a, b):
    n, m = len(a), len(b)
    def check(i, j):
        if i == n:
            return True
        if j == m:
            return False
        if a[i] == b[j]:
            return check(i + 1, j + 1)
        return False
    def try_pos(j):
        if j + n > m:
            return False
        if check(0, j):
            return True
        return try_pos(j + 1)
    return try_pos(0)

a = list(map(int, input().split()))
b = list(map(int, input().split()))
print("YES" if is_subarray(a, b) else "NO")
```

#W9/6

```python
def next_perm(p):
    def find_i(i):
        if i < 0:
            return -1
        if p[i] < p[i + 1]:
            return i
        return find_i(i - 1)
    def find_j(j, i):
        if p[j] > p[i]:
            return j
        return find_j(j - 1, i)
    i = find_i(len(p) - 2)
    if i == -1:
        return sorted(p)
    j = find_j(len(p) - 1, i)
    p[i], p[j] = p[j], p[i]
    p[i + 1:] = reversed(p[i + 1:])
    return p

n = int(input())
p = list(map(int, input().split()))
print(*next_perm(p))
```

#W9/7

```python
def canPlaceFlowers(flowerbed, k):
    def helper(i, k):
        if k == 0:
            return True
        if i >= len(flowerbed):
            return False
        if flowerbed[i] == 0 and (i == 0 or flowerbed[i-1] == 0) and (i == len(flowerbed)-1 or flowerbed[i+1] == 0):
            flowerbed[i] = 1
            return helper(i + 2, k - 1)
        return helper(i + 1, k)
    return helper(0, k)

flowerbed = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

#W9/8

```python
def is_path_crossing(moves):
    def step(i, x, y, seen):
        if (x, y) in seen:
            return True
        if i == len(moves):
            return False
        seen.add((x, y))
        d = moves[i]
        if d == 'R':
            return step(i + 1, x + 1, y, seen)
        if d == 'L':
            return step(i + 1, x - 1, y, seen)
        if d == 'U':
            return step(i + 1, x, y + 1, seen)
        return step(i + 1, x, y - 1, seen)
    return step(0, 0, 0, set())

moves = input().strip()
print(is_path_crossing(moves))
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(float(score))

    def avg(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0.0

    def rank(self):
        average = self.avg()
        if average >= 8: return "Excellent"
        if average >= 6.5: return "Good"
        if average >= 5: return "Average"
        return "Poor"
```

#W9/11

```python
import math

class Point:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __sub__(self, other):
        return Point(self.x - other.x, self.y - other.y, self.z - other.z)

    def cross(self, other):
        return Point(self.y * other.z - self.z * other.y,
                     self.z * other.x - self.x * other.z,
                     self.x * other.y - self.y * other.x)

    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)
```

#W9/12

```python
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = float(realPart)
        self.imaginaryPart = float(imaginaryPart)

    def multiply(self, other):
        new_real = (self.realPart * other.realPart) - (self.imaginaryPart * other.imaginaryPart)
        new_imag = (self.realPart * other.imaginaryPart) + (self.imaginaryPart * other.realPart)
        return ComplexNumber(new_real, new_imag)

    def __str__(self):
        return f"{int(self.realPart)} {int(self.imaginaryPart)}"
```

```python
class Triangle:
    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.edge1 = initEdge1
        self.edge2 = initEdge2
        self.edge3 = initEdge3

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3
        if a + b > c and a + c > b and b + c > a:
            s = (a + b + c) / 2
            area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
            return area
        return -1
```
