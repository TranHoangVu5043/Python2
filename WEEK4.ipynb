#W4/10

```python
def calcNext(n2):
    if n2 % 2 == 0:
        return n2 / 2
    else:
        return n2 * 3 + 1

def collatzLength(n):
    if n == 1: return 0

    step = 1
    curr = n

    while curr != 1:
        curr = calcNext(curr)
        step += 1

    return step

n = int(input())

curr = 0
largest = 0

for i in range(1, n + 1):
    l = collatzLength(i)
    
    if l > largest:
        largest = l
        curr = i
print(curr, largest)
```

#W4/11

```python
n = int(input())

def countDiv(n):
    s = 0
    
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:

            if i % 2 == 0:
                s += 1

            d = n // i

            if d != i and d % 2 == 0: 
                s += 1

    return s

print(countDiv(n))
```

```python
s = input()

if len(s) <= 6:
    print("Weak password")
    exit

u = 0
n = 0
special = 0

for i in s:
    if i == " ": 
        continue

    if i.isalpha():
        u += i.isupper() and 1 or 0

    elif i.isnumeric():
        n += 1
    else:
        special += 1

if (u == 0 or n == 0 or special == 0):
    print("Weak password")
    exit

print("Strong password")
```
