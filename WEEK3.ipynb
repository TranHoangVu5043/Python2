#W3/13

```python
def chiPhiTienDien(kwh):
    if kwh <= 50:
        bill = kwh * 1500
    elif kwh <= 100:
        bill = 50 * 1500 + (kwh - 50) * 2000
    else:
        bill = 50 * 1500 + 50 * 2000 + (kwh - 100) * 3000
    return bill

kwh = float(input("Nhap so kWh dien tieu thu: "))
bill = chiPhiTienDien(kwh)
print(f"Tien dien can thanh toan: {bill} VND")
```
